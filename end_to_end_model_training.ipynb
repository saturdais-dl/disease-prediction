## Arquitecture from paper 


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense


def build_ecg_model():
    model = Sequential()

    # First layer: Convolutional layer with input shape (number of time steps, number of features)
    model.add(Conv1D(filters=24, kernel_size=3, strides=1, padding='same', activation='relu', input_shape=(None, 12)))

    # First Pooling layer
    model.add(MaxPooling1D(pool_size=3, strides=3, padding='valid'))

    # Second Convolutional layer
    model.add(Conv1D(filters=48, kernel_size=3, strides=1, padding='same', activation='relu'))

    # Second Pooling layer
    model.add(MaxPooling1D(pool_size=3, strides=3, padding='valid'))

    # Third Convolutional layer
    model.add(Conv1D(filters=64, kernel_size=3, strides=1, padding='same', activation='relu'))

    # Third Pooling layer
    model.add(MaxPooling1D(pool_size=3, strides=3, padding='valid'))

    # Fourth Convolutional layer
    model.add(Conv1D(filters=72, kernel_size=3, strides=1, padding='same', activation='relu'))

    # Fourth Pooling layer
    model.add(MaxPooling1D(pool_size=3, strides=3, padding='valid'))

    # Fifth Convolutional layer
    model.add(Conv1D(filters=96, kernel_size=3, strides=1, padding='same', activation='relu'))

    # Fifth Pooling layer
    model.add(MaxPooling1D(pool_size=3, strides=3, padding='valid'))

    # Sixth Convolutional layer
    model.add(Conv1D(filters=96, kernel_size=2, strides=1, padding='valid', activation='relu'))

    # Flatten the output to feed into a dense layer
    model.add(Flatten())

    # Fully connected layer
    model.add(Dense(20, activation='relu'))

    return model
```
